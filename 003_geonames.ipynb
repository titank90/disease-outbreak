```python
import re
import pandas as pd
import geonamescache
from unidecode import unidecode

gc = geonamescache.GeonamesCache()
cities = gc.get_cities()


black_lst = ["Can", "Come", "Of"]
city_dict = {}
city_lst = []
city_str = ""

items = cities.items()
for item in items:
    city = item[1]['name']
    countrycode = item[1]['countrycode']
    if city not in black_lst:
        city = unidecode(city)
        city_dict[city] = countrycode

city_lst = [*city_dict]
city_lst.sort(key = len)  
city_lst.reverse()

# print city_lst
city_dict = {k.lower(): v for k, v in city_dict.items()}
# print(city_dict)
city_str = '|'.join(city_lst)
city_str = "\\b(" + city_str + ")\\b"

regexp = re.compile(r"(?P<city>{cities})".format(cities=city_str), re.IGNORECASE)
```

```python
count = 0
cnt_ok = 0
cnt_nok = 0
with open("data/headlines.txt", "r") as fn:
    lines = fn.readlines()
    for line in lines:
        count = count + 1
        line = line.strip()
        result = regexp.search(line)
        if result != None:
            cnt_ok += 1
            city = result.group('city')
            city_hlp = city.lower()
            if city_hlp in city_dict:
                print(count, "--", line, "--", city, "--", city_dict[city_hlp])
            else:
                print(count, "DBG-002:", "--", city, "not found in dictionary")
        else:
            cnt_nok +=1
            print(count, "DBG-001:", "--", line)
```

```python
print("Total",count)   
print("Nr OK",cnt_ok)   
print("Nr NOK",cnt_nok)   
```
